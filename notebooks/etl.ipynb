```python
# Cargamos la hoja Accesos_Por_Tecnologia en un DataFrame y visualizamos su estructura:
Accesos_Por_Tecnologia = excel_file.parse(6)
Accesos_Por_Tecnologia.head(3)
```

```python
# Verificamos si contiene valores nulos (NaNs):
Accesos_Por_Tecnologia.isna().sum()
```

```python
# Procedemos a borrarlos: 
Accesos_Por_Tecnologia.dropna(inplace=True)
```

```python
# Verificamos el tipo de datos que contiene la columna 'Año'
Accesos_Por_Tecnologia['Año'].apply(type).value_counts()
```

```python
# Observamos aquellas filas en la columna 'Año' que sean strings: 
Accesos_Por_Tecnologia[Accesos_Por_Tecnologia['Año'].apply(lambda x: isinstance(x, str))].head(3)
```

```python
# Convertir todos los valores a string para aplicar str.replace a la columna 'Año' entera:
Accesos_Por_Tecnologia['Año'] = Accesos_Por_Tecnologia['Año'].astype(str)

# Reemplazar las filas que contengan el caracter '*' por '': 
Accesos_Por_Tecnologia['Año'] = Accesos_Por_Tecnologia['Año'].str.replace('*', '', regex=False)

# Volver a convertir los valores a enteros:
Accesos_Por_Tecnologia['Año'] = pd.to_numeric(Accesos_Por_Tecnologia['Año'], errors='coerce')
```

```python
# Convertir todos los valores a string para aplicar str.replace a la columna 'Trimestre' entera:
Accesos_Por_Tecnologia['Trimestre'] = Accesos_Por_Tecnologia['Trimestre'].astype(str)

# Reemplazar las filas que contengan el caracter '*' por '': 
Accesos_Por_Tecnologia['Trimestre'] = Accesos_Por_Tecnologia['Trimestre'].str.replace('*', '', regex=False)

# Volver a convertir los valores a enteros:
Accesos_Por_Tecnologia['Trimestre'] = pd.to_numeric(Accesos_Por_Tecnologia['Trimestre'], errors='coerce')
```
